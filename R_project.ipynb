```r
mode<-function(x) {
  o<-na.omit(unique(x))
  tab<-tabulate(match(x,o));o[tab==max(tab)]

}
```

```r
df$Name[is.na(df$Name)]<-mode(df$Name)
df$Genre[is.na(df$Genre)]<-mode(df$Genre)
df$Critic_Score[is.na(df$Critic_Score)]<-mean(df$Critic_Score,na.rm=TRUE)
df$Critic_Count[is.na(df$Critic_Count)]<-mean(df$Critic_Count,na.rm=TRUE)
df$User_Count[is.na(df$User_Count)]<-mean(df$User_Count,na.rm=TRUE)
df$User_Score[is.na(df$User_Score)]<-mode(df$User_Score)
df$Developer[is.na(df$Developer)]<-mode(df$Developer)
df$Rating[is.na(df$Rating)]<-mode(df$Rating)

colSums(is.na(df))
```

```r
Genre<-df %>%
  group_by(Genre) %>%
  summarize(count=n()) %>%
  arrange(count)
Genre
```

```r
ggplot(Genre,aes(x="",y=count,fill=Genre))+geom_col()+coord_polar(theta="y")+geom_text(aes(label=count),position = position_stack(vjust = 0.5))
```

```r
Platform<-df %>%
  group_by(Platform) %>%
  summarize(count=n()) %>%
  arrange(count)

ggplot(Platform,aes(x=Platform,y=count,fill=Platform))+geom_bar(stat="identity")+
       labs(title="Platform Popularity",x="Platform")
```

```r
df %>%
  group_by(Genre,Year_of_Release) %>%
  summarize(total_sales=sum(Global_Sales)) %>%
  ggplot(aes(x=Year_of_Release,y=total_sales,fill=Genre))+ geom_col() +scale_fill_brewer(palette = "Set3")+
       labs(title="Genre Popularity over the year by sales",x="Year",y="Total_Sales")
```

```r
df$Year_of_Release<-as.numeric(df$Year_of_Release)
```

```r
df %>%
  subset(select=c(Year_of_Release,NA_Sales,EU_Sales,JP_Sales,Other_Sales,Global_Sales)) %>%
  group_by(Year_of_Release)%>%
  summarize(Na=sum(NA_Sales),Eu=sum(EU_Sales),Jp=sum(JP_Sales),Ot=sum(Other_Sales),Gs=sum(Global_Sales)) %>%
  ggplot()+geom_line(aes(x=Year_of_Release,y=Na,colour="Na"),size=1)+geom_line(aes(x=Year_of_Release,y=Eu,colour="Eu"),size=1)+geom_line(aes(x=Year_of_Release,y=Jp,colour="Jp"),size=1)+
  geom_line(aes(x=Year_of_Release,y=Ot,colour="Ot"),size=1)+labs(title = "Sales Trend over the years", y = "Sales(Millions)")
```
